```python
import pandas as pd
import requests
import json

```

```python
url = 'http://jservice.io/api/clues'
r = requests.get(url)
x = r.json()
df = pd.read_json(json.dumps(x))
```

```python
df[['question','answer','created_at','updated_at']].style.format({"updated_at": lambda t: t.strftime("%d-%m-%Y"), "created_at": lambda t: t.strftime("%d-%m-%Y")})
```
